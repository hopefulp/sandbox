# VaspForces script Demo
VaspForces.sh is a bash script allows to extract forces and atoms position from VASP OUTCAR.<br>
[Script source](https://github.com/K4ys4r/VASP_Scripts/tree/master/Extract_Forces/VaspForces.sh)

```python
%%bash
./VaspForces.sh -infos
```

```python
%%bash
./VaspForces.sh -h
```

```python
%%bash
./VaspForces.sh -outcar=OUTCAR_test -iter=1 
```

```python
%%bash
./VaspForces.sh -outcar=OUTCAR_test -iter=1 -atom=4
```

```python
%%bash
./VaspForces.sh -outcar=OUTCAR_test -iter=all -atom=4
```

```python
%%bash
./VaspForces.sh -outcar=OUTCAR_test -iter=all -atom=all
```
